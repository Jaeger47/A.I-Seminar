## 2. Take a peek at your raw data.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

names = ['RC', 'SA', 'WA', 'RA', 'OH', 'O', 'GA', 'GAD', 'HL', 'CL']
data = pd.read_csv('Energy_efficiency_DataSet.csv', names=names, comment='#')
peek = data.head(20)
print(peek)
```

## 3. Retrieve `features` and `class`

```python
array = data.values
X = array[:,0:-2]
Y = array[:,-2]
```

```python
print(X)  # features
```

```python
print(Y) # class
```

## 4. Prepare the best `model`

```python
from sklearn.linear_model import LinearRegression

best_model = LinearRegression(copy_X=True, fit_intercept=True, n_jobs=None, normalize=False)
```

## 5. `Split` dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=7)
```

## 6. `Train` best model

```python
best_model.fit(X_train, Y_train)
```

## 7. Results

```python
from sklearn.metrics import r2_score

print(r2_score(Y_test, best_model.predict(X_test)))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(Y_test, best_model.predict(X_test))
plt.show()
```
