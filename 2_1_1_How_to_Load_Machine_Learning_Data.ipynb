# Three (3) Ways to load Machine Learning Data into Google Colab

Importing your data to a notebook is the best way to start your machine learning journey. The most common format for predictive modeling is CSV files. In this notebook, I will be sharing the three ways to load a CSV file in Google Colab. You might learn some alternative way to do this task from the Internet. But in this lesson, I will be teaching to you the mostly used approaches. To be exact, you will learn in this notebook the following:

## 1. Uploading the data from Local File System

You can easily upload a CSV file from your local file system to Google Colab. 

```python
from google.colab import files

uploaded = files.upload()
```

After running the previous line, you will then see the file under Files tab in this notebook. Meaning, the said file is uploaded and it now exists in the directory. Thus, you only need to upload the data once. You can comment the previous code snippet to exclude it from the next run.

Now, you can load the data using the Pandas library and the `pandas.read_csv() `function. 

```python
filename = 'pima-indians-diabetes.csv'
```

```python
import pandas as pd 

names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
data = pd.read_csv(filename, names=names, comment='#')  # I am using the `filename` variable here.
print(data.head())  
```

## 2. Importing the data from Github

Github is a code hosting platform for version control and collaboration among software developers. It is an open source community.

If you happen to have a Github account or interested to have one, you can upload your dataset (`< 25MB`) in there and use the following code below to access them smoothly.
```python
url = 'https://raw.githubusercontent.com/Jaeger47/A.I-Seminar/main/Energy_efficiency_DataSet.csv'
```

After providing the Github link, we will load the data using again the `pd.read_csv()` function of the Pandas library.

```python
import pandas as pd 
#X1 Relative Compactness
#X2 Surface Area
#X3 Wall Area
#X4 Roof Area
#X5 Overall Height
#X6 Orientation
#X7 Glazing Area
#X8 Glazing Area Distribution
#y1 Heating Load
#y2 Cooling Load
names = ['RC', 'SA', 'WA', 'RA', 'OH', 'O', 'GA', 'GAD', 'HL', 'CL']
data = pd.read_csv(url, names=names, comment='#')   # I am using the `url` variable here.
print(data.head())
```

## 3. Importing the data from Google Drive


One of the good practices is to upload your dataset in your Google Drive especially when the dataset is large in size and you do not want it to be available in public due to some non-disclosure agreements. You can perform this task in two ways.

### 3.1. Mounting your Google Drive in the runtime's virtual machine (If dataset is saved in your own Google Drive)

To connect to Google Drive, you need to execute the following two lines. It will then ask you for an authorization code. What you need to do is to sign in your Google accound and copy the authorization code.

```python
from google.colab import drive
drive.mount('/content/drive')
```

Now, you need to provide the location of your dataset.

```python
location = '/content/drive/MyDrive/DATASET/Energy_efficiency_DataSet.csv'
```

We will use again the `pd.read_csv()` function of Pandas library to load the dataset.

```python
import pandas as pd
#X1 Relative Compactness
#X2 Surface Area
#X3 Wall Area
#X4 Roof Area
#X5 Overall Height
#X6 Orientation
#X7 Glazing Area
#X8 Glazing Area Distribution
#y1 Heating Load
#y2 Cooling Load
names = ['RC', 'SA', 'WA', 'RA', 'OH', 'O', 'GA', 'GAD', 'HL', 'CL']
data = pd.read_csv(location, names=names, comment='#')  # I am using the `location` variable here.
print(data.head())
```

### 3.2. Using a wrapper around the API such as [PyDrive](https://pythonhosted.org/PyDrive/) (Dataset may or may not be in your own Google Drive)

Import necessary libraries for authentication using PyDrive wrapper.

```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

Authenticate and create the PyDrive client.

```python
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

Once you have completed the verification process, go to the CSV file in Google Drive. Right-click on it and select **Get link**. The sharable link contains the `id` of the file. Copy it and paste it in the variable `id` below.
